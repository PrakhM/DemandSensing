```python
import pandas as pd
import numpy as np
import random

random.seed(42)
np.random.seed(42)

N_ROWS = 2000

cities = [
    "Mumbai",
    "Delhi",
    "Bengaluru",
    "Hyderabad",
    "Pune",
    "Lucknow",
    "Bhopal"
]

channels = [
    "Myntra",
    "Offline",
    "Brand Website"
]

subcategories = [
    "Kurta",
    "Saree",
    "Lehenga"
]

weather_types = [
    "Summer",
    "Monsoon",
    "Winter"
]

festival_months = [10, 11] 

wedding_months = [2, 11, 12]

city_factor = {
    "Mumbai": 1.30,
    "Delhi": 1.35,
    "Bengaluru": 1.10,
    "Hyderabad": 1.15,
    "Pune": 1.05,
    "Lucknow": 0.95,
    "Bhopal": 0.90
}

channel_factor = {
    "Myntra": 1.20,
    "Offline": 1.0,
    "Brand Website": 1.10
}

dates = pd.date_range(
    start="2023-01-01",
    end="2025-12-31"
)

rows = []

skuDone = {1}

for i in range(N_ROWS):

    sku_id = f"SKU{random.randint(1000,9999)}"
    while(sku_id in skuDone):
        sku_id = f"SKU{random.randint(1000,9999)}"

    skuDone.add(sku_id)

    city = random.choice(cities)

    channel = random.choice(channels)

    subcategory = random.choice(subcategories)

    date = random.choice(dates)

    month = date.month

    if month in [3,4,5,6]:
        weather = "Summer"

    elif month in [7,8,9]:
        weather = "Monsoon"

    else:
        weather = "Winter"

    base_price_map = {
        "Kurta": 2500,
        "Saree": 5000,
        "Lehenga": 9000
    }

    price = (
        base_price_map[subcategory]
        + random.randint(-1000, 1500)
    )

    price = max(1500, price)

    discount = random.choice(
        [0,0,10,15,20,25,30]
    )

    rating = round(
        random.uniform(3.5,5.0),
        1
    )

    sales = 120

    sales *= city_factor[city]

    sales *= channel_factor[channel]

    sales *= (
        1 + discount/100
    )

    sales *= (
        rating / 4
    )

    price_penalty = max(
        0.4,
        1 - (price/15000)
    )

    sales *= price_penalty

    if weather == "Summer":
        if subcategory in [
            "Lehenga"
        ]:
            sales *= 0.70
        if subcategory == "Kurta":
            sales *= 1.10
    elif weather == "Monsoon":
        if channel == "Offline":
            sales *= 0.80
        if subcategory in [
            "Lehenga"
        ]:
            sales *= 0.75
    elif weather == "Winter":
        if subcategory in [
            "Lehenga",
            "Saree"
        ]:
            sales *= 1.25
        if subcategory in "Kurta":
            sales *= 0.8

    if month in festival_months:
        sales *= 1.75

    if month in wedding_months:
        if subcategory in [
            "Lehenga",
            "Saree"
        ]:
            sales *= 1.60

    sales *= random.uniform(
        0.75,
        1.25
    )

    sales = int(max(5, sales))

    rows.append({

        "date": date,

        "sku_id": sku_id,

        "category": "Ethnic Wear",

        "subcategory": subcategory,

        "price": price,

        "discount_percentage": discount,

        "rating": rating,

        "city": city,

        "channel": channel,

        "weather": weather,

        "90_days_sales": sales
    })


df = pd.DataFrame(rows)
```
