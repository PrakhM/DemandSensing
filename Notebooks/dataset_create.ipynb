```python
import pandas as pd
import numpy as np
import random

random.seed(42)
np.random.seed(42)

N_ROWS = 20000

cities = [
    "Mumbai",
    "Delhi",
    "Bengaluru",
    "Hyderabad",
    "Pune",
    "Lucknow",
    "Bhopal"
]

channels = [
    "Myntra",
    "Offline",
    "Brand Website"
]

subcategories = [
    "Kurta",
    "Sherwani",
    "Saree",
    "Lehenga",
    "Nehru Jacket"
]

weather_types = [
    "Summer",
    "Monsoon",
    "Winter"
]

festival_months = [10, 11] 

wedding_months = [2, 11, 12]

city_factor = {
    "Mumbai": 1.30,
    "Delhi": 1.35,
    "Bengaluru": 1.10,
    "Hyderabad": 1.15,
    "Pune": 1.05,
    "Lucknow": 0.95,
    "Bhopal": 0.90
}

channel_factor = {
    "Myntra": 1.20,
    "Offline": 1.0,
    "Brand Website": 1.10
}

dates = pd.date_range(
    start="2024-01-01",
    end="2025-12-31"
)

rows = []

for i in range(N_ROWS):

    sku_id = f"SKU{random.randint(1000,9999)}"

    city = random.choice(cities)

    channel = random.choice(channels)

    subcategory = random.choice(subcategories)

    date = random.choice(dates)

    month = date.month

    if month in [3,4,5,6]:
        weather = "Summer"

    elif month in [7,8,9]:
        weather = "Monsoon"

    else:
        weather = "Winter"

    base_price_map = {
        "Kurta": 2500,
        "Sherwani": 8000,
        "Saree": 5000,
        "Lehenga": 9000,
        "Nehru Jacket": 4000
    }

    price = (
        base_price_map[subcategory]
        + random.randint(-1000, 1500)
    )

    price = max(1500, price)

    discount = random.choice(
        [0,0,10,15,20,25,30]
    )

    rating = round(
        random.uniform(3.5,5.0),
        1
    )

    review_count = random.randint(
        50,
        8000
    )
    

    sales = 120

    sales *= city_factor[city]

    sales *= channel_factor[channel]

    sales *= (
        1 + discount/100
    )

    sales *= (
        rating / 4
    )

    sales *= (
        1 + np.log1p(review_count)/12
    )

    price_penalty = max(
        0.4,
        1 - (price/15000)
    )

    sales *= price_penalty

    if weather == "Summer":

        if subcategory in [
            "Sherwani",
            "Lehenga"
        ]:
            sales *= 0.70

        if subcategory == "Kurta":
            sales *= 1.10

    elif weather == "Monsoon":
        if channel == "Offline":
            sales *= 0.80
        if subcategory in [
            "Sherwani",
            "Lehenga"
        ]:
            sales *= 0.75

    elif weather == "Winter":

        if subcategory in [
            "Sherwani",
            "Lehenga",
            "Saree"
        ]:
            sales *= 1.25

    if month in festival_months:

        sales *= 1.50

        if subcategory in [
            "Saree",
            "Lehenga",
            "Kurta"
        ]:
            sales *= 1.20

    if month in wedding_months:

        if subcategory in [
            "Sherwani",
            "Lehenga",
            "Saree"
        ]:
            sales *= 1.60

    sales *= random.uniform(
        0.75,
        1.25
    )

    stockout = np.random.choice(
        [0,1],
        p=[0.96,0.04]
    )

    if stockout == 1:
        sales *= 0.50

    sales = int(max(5, sales))

    rows.append({

        "date": date,

        "sku_id": sku_id,

        "category": "Ethnic Wear",

        "subcategory": subcategory,

        "price": price,

        "discount_percentage": discount,

        "rating": rating,

        "review_count": review_count,

        "city": city,

        "channel": channel,

        "weather": weather,

        "stockout": stockout,

        "90_days_sales": sales
    })


df = pd.DataFrame(rows)

print(df.head())

print("\nShape:", df.shape)

df.to_csv(
    "ethnic_wear_historical_data.csv",
    index=False
)

print("\nDataset saved successfully.")
```
