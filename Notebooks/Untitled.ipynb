```python
import random
import pandas as pd
import numpy as np

cities = [
    "Mumbai",
    "Delhi",
    "Bengaluru",
    "Hyderabad",
    "Lucknow",
    "Pune",
    "Prayagraj",
    "Bhopal"
]

age_groups = [
    "22-28",
    "29-35",
    "36-38"
]

channels = [
    "Myntra",
    "Offline",
    "Brand website"
]

discover_channels = [
    "Instagram",
    "Influencer",
    "Friend",
    "Myntra browse",
    "Other"
]

events = [
    "Wedding",
    "Festival",
    "Work event",
    "Gifting",
    "NA"
]

fair_price = [
    2500,
    3500,
    4500,
    5500,
    6500
]

city_multiplier = {
    "Mumbai": 1.3,
    "Delhi": 1.25,
    "Bengaluru": 1.15,
    "Hyderabad": 1.05,
    "Pune": 1.1,
    "Lucknow": 0.9,
    "Prayagraj": 0.8,
    "Bhopal": 0.85
}

age_multiplier = {
    "22-28": 1.2,
    "29-35": 1.0,
    "36-38": 0.9
}
```

```python
rows = []

NUM_RESPONSES = 1000

for i in range(NUM_RESPONSES):

    respondent_id = f"R{i+1:04}"

    city = random.choice(cities)
    age_group = random.choice(age_groups)

    preferred_channel = random.choice(channels)

    discover_channel = random.choice(discover_channels)

    event = random.choice(events)

    base_intent = random.randint(3, 10)

    base_intent *= city_multiplier[city]
    base_intent *= age_multiplier[age_group]
    if discover_channel == "Influencer":
        base_intent += 1
    if event in ["Wedding", "Festival"]:
        base_intent += 1
    purchase_intent = int(np.clip(round(base_intent), 1, 10))

    price_acceptance_index = int(np.clip(purchase_intent + random.randint(-1, 1),1,10)/2.1)

    price_acceptance = fair_price[price_acceptance_index]

    rows.append({
        "respondent_id": respondent_id,
        "city": city,
        "age_group": age_group,
        "preferred_channel": preferred_channel,
        "discover_channel": discover_channel,
        "primary_occasion": event,
        "purchase_intent" : purchase_intent,
        "price_acceptance": price_acceptance
    })

df = pd.DataFrame(rows)
```

```python
from pathlib import Path

notebook_dir = Path.cwd() 
data_dir = notebook_dir.parent / 'Data'
save_path = data_dir / 'survery.csv'

df.to_csv(
    save_path,
    index=False
)
```
